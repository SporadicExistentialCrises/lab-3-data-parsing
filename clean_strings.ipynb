```python
# Task B
import pandas as pd
import string

def clean_strings(strings):
    clean_string_list = []
    for word in strings:
        # Remove leading and trailing whitespace
        word = word.strip()
        # Convert to lowercase
        word = word.lower()
        # Remove punctuation and spaces
        word = word.replace(' ', '')
        word = word.translate(str.maketrans('', '', string.punctuation))
        if word:
            # If the string is not empty, keep it
            clean_string_list.append(word)
    
    return clean_string_list

# Example usage
print(clean_strings(['  Hello, World!  ','    ','Python is great.', 'Data Science 101!', '']))
```

```python
# Task D
import pandas as pd
import string

def clean_strings(strings):
    clean_string_list = []
    for word in strings:
        # Remove leading and trailing whitespace
        word = word.strip()
        # Convert to lowercase
        word = word.lower()
        # Remove punctuation and spaces
        word = word.replace(' ', '')
        word = word.translate(str.maketrans('', '', string.punctuation))
        if word:
            # If the string is not empty, keep it
            clean_string_list.append(word)
    
    return clean_string_list

df = pd.read_csv('messy_strings.csv')
df["clean"] = clean_strings(df["raw"]) 
print(f"List size: {len(df['clean'])}")
unique_words = set(df["clean"])
print(f"Unique words: {len(unique_words)}")
words_count = df["clean"].value_counts()
print(f"Most common word: {words_count.idxmax()}")
```
